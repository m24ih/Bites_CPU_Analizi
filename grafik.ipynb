```python
import pandas as pd
import os
import plotly.express as px
```

```python
def analyze_all_metrics(processed_data_path):
    """
    Tm metrikleri ieren ilenmi veriyi okur ve grselletirir.
    """
    if processed_data_path is None or not os.path.exists(processed_data_path):
        print("\nlenecek veri dosyas bulunamad. Analiz adm atlanyor.")
        return

    print("\n--- ADIM 2: Tm Metriklerin Analizi ve Grselletirilmesi ---")
    
    # 1. lenmi Veriyi Oku
    df = pd.read_csv(processed_data_path, index_col='timestamp')
    df.index = pd.to_datetime(df.index, unit='s')
    print("lenmi veri okundu.")
    print(df.head())
    
    # 2. Veriyi Grselletir
    print("Tm genel metriklerin grafii oluturuluyor...")
    fig = px.line(
        df, 
        y=['avg_min_cpu', 'avg_max_cpu', 'avg_avg_cpu'],
        title='Zaman inde Sistem Genelindeki Ortalama CPU Metrikleri',
        labels={'value': 'Ortalama CPU Yzdesi', 'timestamp': 'Zaman', 'variable': 'Metrik'}
    )
    fig.update_layout(legend_title_text='Ortalama Metrikler')
    fig.show()
    
    print("\n--- ADIM 2 BAARIYLA TAMAMLANDI! ---")
    print("Grafik, sistemin genel yknn alt ve st snrlarn gstermektedir.")
```

```python

output_file = '/home/melih/Documents/coding/Bites_CPU_Analizi/genel_sunucu_yuku_tum_metrikler.csv'
analyze_all_metrics(output_file)
```
