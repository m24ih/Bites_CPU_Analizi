```python
import pandas as pd
import os
import plotly.express as px
```

```python
def analyze_all_metrics(processed_data_path):
    """
    Tüm metrikleri içeren işlenmiş veriyi okur ve görselleştirir.
    """
    if processed_data_path is None or not os.path.exists(processed_data_path):
        print("\nİşlenecek veri dosyası bulunamadı. Analiz adımı atlanıyor.")
        return

    print("\n--- ADIM 2: Tüm Metriklerin Analizi ve Görselleştirilmesi ---")
    
    # 1. İşlenmiş Veriyi Oku
    df = pd.read_csv(processed_data_path, index_col='timestamp')
    df.index = pd.to_datetime(df.index, unit='s')
    print("İşlenmiş veri okundu.")
    print(df.head())
    
    # 2. Veriyi Görselleştir
    print("Tüm genel metriklerin grafiği oluşturuluyor...")
    fig = px.line(
        df, 
        y=['avg_min_cpu', 'avg_max_cpu', 'avg_avg_cpu'],
        title='Zaman İçinde Sistem Genelindeki Ortalama CPU Metrikleri',
        labels={'value': 'Ortalama CPU Yüzdesi', 'timestamp': 'Zaman', 'variable': 'Metrik'}
    )
    fig.update_layout(legend_title_text='Ortalama Metrikler')
    fig.show()
    
    print("\n--- ADIM 2 BAŞARIYLA TAMAMLANDI! ---")
    print("Grafik, sistemin genel yükünün alt ve üst sınırlarını göstermektedir.")
```

```python

output_file = '/home/melih/Documents/coding/Bites_CPU_Analizi/genel_sunucu_yuku_tum_metrikler.csv'
analyze_all_metrics(output_file)
```
